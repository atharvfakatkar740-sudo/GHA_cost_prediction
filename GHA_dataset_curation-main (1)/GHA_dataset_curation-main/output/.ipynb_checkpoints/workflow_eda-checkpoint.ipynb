````python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
import seaborn as sns
from scipy import stats
from scipy.stats import (pearsonr, spearmanr, kendalltau, shapiro,
                          kruskal, mannwhitneyu, pointbiserialr)
from scipy.cluster import hierarchy
from scipy.spatial.distance import squareform
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import mutual_info_regression, f_regression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.inspection import permutation_importance
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from IPython.display import Markdown, display

plt.rcParams.update({
    'figure.dpi': 130,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'axes.grid': True,
    'grid.alpha': 0.35,
    'font.size': 11,
    'axes.titlesize': 13,
    'axes.titleweight': 'bold',
})
PALETTE  = sns.color_palette('tab10')
TARGET   = 'duration_minutes'
SEED     = 42
print('Libraries loaded OK')
````

## 1 · Dataset Overview & Schema Audit

````python
df_raw = pd.read_csv('workflow_features.csv')
print(f'Shape: {df_raw.shape[0]} rows x {df_raw.shape[1]} columns')
df_raw.head()
````

````python
df = df_raw.copy()

bool_cols = ['uses_matrix_strategy','fail_fast','is_using_setup_actions',
             'is_using_docker_actions','is_using_cache']
for c in bool_cols:
    df[c] = df[c].astype(str).str.lower().map({'true': True, 'false': False})
df['container_image'] = df['container_image'].astype(str).str.lower().apply(
    lambda x: False if x == 'false' else True)
bool_cols.append('container_image')

num_cols = ['total_cost_usd','duration_minutes','yaml_line_count','yaml_depth',
            'job_count','total_steps','avg_steps_per_job','matrix_dimensions',
            'matrix_permutations','timeout_minutes','unique_actions_used',
            'env_var_count','if_condition_count','needs_dependencies_count','code_complexity']
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')

CAT_COLS  = ['os_label','primary_language','workflow_name']
BOOL_COLS = bool_cols
NUM_COLS  = [c for c in num_cols if c not in [TARGET,'total_cost_usd']]

print(f'Numeric features : {len(NUM_COLS)}')
print(f'Boolean features : {len(BOOL_COLS)}')
miss = df.isnull().sum()
print(f'Missing values   : {miss[miss>0].to_dict() or "none"}')
````

````python
# Fig 1 — Schema audit: missingness map + cardinality
fig, axes = plt.subplots(1, 2, figsize=(18, 5))
fig.suptitle('Fig 1 — Dataset Schema Audit', fontsize=14, fontweight='bold')

miss_map = df.isnull().astype(int)
sns.heatmap(miss_map.T, cmap='Reds', cbar=False, linewidths=0.3,
            yticklabels=True, ax=axes[0])
axes[0].set_title('Missing-Value Map (red = missing)')
axes[0].set_xlabel('Row index')

cardinality = df.nunique().sort_values(ascending=True)
cmap_colors = ['#e74c3c' if c == TARGET else
               '#3498db' if c in NUM_COLS else
               '#2ecc71' if c in BOOL_COLS else '#f39c12'
               for c in cardinality.index]
axes[1].barh(cardinality.index, cardinality.values, color=cmap_colors, edgecolor='white')
axes[1].set_title('Unique Values per Column (cardinality)')
axes[1].set_xlabel('Unique values')
patches = [mpatches.Patch(color='#e74c3c', label='Target'),
           mpatches.Patch(color='#3498db', label='Numeric'),
           mpatches.Patch(color='#2ecc71', label='Boolean'),
           mpatches.Patch(color='#f39c12', label='Categorical/ID')]
axes[1].legend(handles=patches)
plt.tight_layout()
plt.show()

df.describe(include='all').T
````

## 2 · Target Variable Deep Dive — `duration_minutes`

````python
y = df[TARGET]

# Fig 2 — Distribution: histogram + KDE + ECDF + Q-Q
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Fig 2 — Target Variable: duration_minutes', fontsize=14, fontweight='bold')

axes[0,0].hist(y, bins=30, color=PALETTE[0], edgecolor='white', alpha=0.8, density=True)
y.plot.kde(ax=axes[0,0], color='black', lw=2)
axes[0,0].axvline(y.mean(),   color='red',    ls='--', lw=1.5, label=f'Mean={y.mean():.2f}')
axes[0,0].axvline(y.median(), color='orange', ls='--', lw=1.5, label=f'Median={y.median():.2f}')
axes[0,0].set_title('Histogram + KDE')
axes[0,0].legend()

y_log = np.log1p(y)
axes[0,1].hist(y_log, bins=25, color=PALETTE[1], edgecolor='white', alpha=0.8, density=True)
y_log.plot.kde(ax=axes[0,1], color='black', lw=2)
axes[0,1].set_title('log1p(duration) — Testing for Normality')
axes[0,1].set_xlabel('log1p(duration_minutes)')

sorted_y = np.sort(y)
ecdf = np.arange(1, len(sorted_y)+1) / len(sorted_y)
axes[1,0].plot(sorted_y, ecdf, lw=2, color=PALETTE[2])
axes[1,0].fill_between(sorted_y, ecdf, alpha=0.15, color=PALETTE[2])
for pct in [0.25, 0.5, 0.75, 0.9]:
    val = np.quantile(sorted_y, pct)
    axes[1,0].axvline(val, ls=':', color='gray', lw=1)
    axes[1,0].text(val+0.3, pct-0.06, f'P{int(pct*100)}={val:.1f}', fontsize=8)
axes[1,0].set_title('Empirical CDF')
axes[1,0].set_xlabel('duration_minutes')
axes[1,0].set_ylabel('Cumulative probability')

stats.probplot(y, dist='norm', plot=axes[1,1])
axes[1,1].set_title('Q-Q Plot vs Normal Distribution')
plt.tight_layout()
plt.show()

stat, p_sw = shapiro(y)
print(f'Shapiro-Wilk: W={stat:.4f}, p={p_sw:.4e} -> {"NOT normal" if p_sw<0.05 else "Normal"}')
print(f'Skewness={y.skew():.3f}  Kurtosis={y.kurtosis():.3f}')
````

````python
# Fig 3 — Duration by workflow name
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
fig.suptitle('Fig 3 — Duration by Workflow Name', fontsize=14, fontweight='bold')

wf_order = df.groupby('workflow_name')[TARGET].median().sort_values(ascending=False).index
short_names = {w: (w[:38]+'...' if len(w)>38 else w) for w in wf_order}
df['wf_short'] = df['workflow_name'].map(short_names)
short_order = [short_names[w] for w in wf_order]

sns.boxplot(data=df, y='wf_short', x=TARGET, order=short_order, palette='tab10', ax=axes[0])
sns.stripplot(data=df, y='wf_short', x=TARGET, order=short_order, color='black',
              alpha=0.4, jitter=True, size=4, ax=axes[0])
axes[0].set_title('Box + Strip: Median and Spread')
axes[0].set_ylabel('')

sns.violinplot(data=df, y='wf_short', x=TARGET, order=short_order,
               palette='tab10', inner='quartile', ax=axes[1])
axes[1].set_title('Violin: Full Distribution Shape')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()

groups = [g[TARGET].values for _, g in df.groupby('workflow_name')]
H, p_kw = kruskal(*groups)
print(f'Kruskal-Wallis (workflows): H={H:.2f}, p={p_kw:.2e}')
print(f'  -> Workflows differ {"significantly" if p_kw<0.05 else "non-significantly"} in duration')
````

````python
# Fig 4 — Duration by OS runner
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Fig 4 — Duration by OS Runner', fontsize=14, fontweight='bold')

os_order = df.groupby('os_label')[TARGET].median().sort_values(ascending=False).index

sns.boxplot(data=df, x='os_label', y=TARGET, order=os_order, palette='Set2', width=0.4, ax=axes[0])
sns.stripplot(data=df, x='os_label', y=TARGET, order=os_order, color='black',
              alpha=0.4, jitter=True, size=5, ax=axes[0])
axes[0].set_title('Box + Strip')

os_stats = df.groupby('os_label')[TARGET].agg(['mean','sem']).reindex(os_order)
axes[1].bar(os_stats.index, os_stats['mean'], yerr=1.96*os_stats['sem'],
            color=sns.color_palette('Set2', len(os_stats)), capsize=6, edgecolor='white')
axes[1].set_title('Mean +/- 95% CI')
axes[1].set_ylabel('duration_minutes')

sns.swarmplot(data=df, x='os_label', y=TARGET, order=os_order, palette='Set2', size=6, ax=axes[2])
axes[2].set_title('Swarm Plot (every data point)')

plt.tight_layout()
plt.show()

if df['os_label'].nunique() == 2:
    cats = df['os_label'].unique()
    g1 = df[df['os_label']==cats[0]][TARGET]
    g2 = df[df['os_label']==cats[1]][TARGET]
    u, p_mw = mannwhitneyu(g1, g2, alternative='two-sided')
    r_eff = 1 - (2*u)/(len(g1)*len(g2))
    print(f'Mann-Whitney U ({cats[0]} vs {cats[1]}): U={u:.0f}, p={p_mw:.4f}, effect r={r_eff:.3f}')
````

## 3 · Univariate Feature Distributions

````python
# Fig 5 — Histograms: all numeric features
n = len(NUM_COLS)
ncols = 4; nrows = -(-n // ncols)
fig, axes = plt.subplots(nrows, ncols, figsize=(20, nrows*3.5))
fig.suptitle('Fig 5 — Univariate Distributions: All Numeric Features', fontsize=14, fontweight='bold')
axes = axes.flatten()

for i, col in enumerate(NUM_COLS):
    data = df[col].dropna()
    axes[i].hist(data, bins=20, color=PALETTE[i % len(PALETTE)], edgecolor='white', alpha=0.85, density=True)
    if data.nunique() > 5:
        try: data.plot.kde(ax=axes[i], color='black', lw=1.5)
        except: pass
    axes[i].set_title(col)
    axes[i].text(0.97, 0.93, f'skew={data.skew():.2f}', transform=axes[i].transAxes,
                 ha='right', va='top', fontsize=8, color='darkred')

for j in range(i+1, len(axes)): axes[j].set_visible(False)
plt.tight_layout()
plt.show()
````

````python
# Fig 6 — Boolean feature prevalence
fig, axes = plt.subplots(2, 4, figsize=(20, 9))
fig.suptitle('Fig 6 — Boolean Feature Prevalence', fontsize=14, fontweight='bold')
axes = axes.flatten()

for i, col in enumerate(BOOL_COLS[:8]):
    vc = df[col].value_counts()
    colors = ['#2ecc71' if str(v)=='True' else '#e74c3c' for v in vc.index]
    axes[i].bar([str(v) for v in vc.index], vc.values, color=colors, edgecolor='white', width=0.5)
    axes[i].set_title(col)
    for bar, val in zip(axes[i].patches, vc.values):
        axes[i].text(bar.get_x()+bar.get_width()/2, bar.get_height()+0.3, str(val), ha='center')

plt.tight_layout(); plt.show()

# Categorical value counts
for col in ['os_label', 'workflow_name']:
    vc = df[col].value_counts()
    fig, ax = plt.subplots(figsize=(max(8, len(vc)*2), 4))
    ax.bar(range(len(vc)), vc.values, color=sns.color_palette('tab10', len(vc)), edgecolor='white')
    ax.set_xticks(range(len(vc)))
    ax.set_xticklabels([str(v)[:40] for v in vc.index], rotation=30, ha='right', fontsize=9)
    ax.set_title(f'Fig 7 — Value Counts: {col}', fontweight='bold')
    for bar, val in zip(ax.patches, vc.values):
        ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+0.1, str(val), ha='center', fontsize=9)
    plt.tight_layout(); plt.show()
````

## 4 · Bivariate Analysis — Each Feature vs `duration_minutes`

````python
# Fig 8 — Scatter plots: numeric features vs target, coloured by OS
n = len(NUM_COLS)
ncols = 4; nrows = -(-n // ncols)
fig, axes = plt.subplots(nrows, ncols, figsize=(20, nrows*4))
fig.suptitle('Fig 8 — Scatter Plots: Numeric Features vs duration_minutes', fontsize=14, fontweight='bold')
axes = axes.flatten()

os_palette = {'ubuntu': '#3498db', 'self-hosted': '#e67e22', 'windows': '#9b59b6', 'macos': '#1abc9c'}

for i, col in enumerate(NUM_COLS):
    x_ = df[col].dropna()
    y_ = df.loc[x_.index, TARGET]
    c_ = df.loc[x_.index, 'os_label'].map(os_palette).fillna('#95a5a6')
    axes[i].scatter(x_, y_, c=c_, alpha=0.6, s=35, edgecolors='white', lw=0.5)
    if x_.nunique() > 2:
        m, b, r, p_lr, _ = stats.linregress(x_, y_)
        xs = np.linspace(x_.min(), x_.max(), 100)
        axes[i].plot(xs, m*xs+b, 'r-', lw=1.5, alpha=0.7)
        r_col = '#c0392b' if abs(r)>0.5 else '#27ae60' if abs(r)<0.2 else '#e67e22'
        axes[i].set_title(f'{col}\nr={r:.3f}, p={p_lr:.3f}', color=r_col, fontsize=10)
    else:
        axes[i].set_title(col)
    axes[i].set_xlabel(col, fontsize=8)
    axes[i].set_ylabel(TARGET, fontsize=8)

for j in range(i+1, len(axes)): axes[j].set_visible(False)
fig.legend(handles=[mpatches.Patch(color=v, label=k) for k,v in os_palette.items()],
           title='OS Runner', loc='lower right')
plt.tight_layout(); plt.show()
````

````python
# Fig 10 — Boolean features: box + jitter vs target
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
fig.suptitle('Fig 10 — Boolean Features vs duration_minutes', fontsize=14, fontweight='bold')
axes = axes.flatten()

for i, col in enumerate(BOOL_COLS):
    tmp = df[[col, TARGET]].dropna().copy()
    tmp[col] = tmp[col].astype(str)
    sns.boxplot(data=tmp, x=col, y=TARGET, palette=['#e74c3c','#2ecc71'],
                order=['False','True'], ax=axes[i], width=0.4)
    sns.stripplot(data=tmp, x=col, y=TARGET, order=['False','True'],
                  color='black', alpha=0.4, jitter=True, size=4, ax=axes[i])
    g_f = tmp[tmp[col]=='False'][TARGET]
    g_t = tmp[tmp[col]=='True'][TARGET]
    if len(g_f)>1 and len(g_t)>1:
        _, p_mw = mannwhitneyu(g_f, g_t, alternative='two-sided')
        sig = '***' if p_mw<0.001 else '**' if p_mw<0.01 else '*' if p_mw<0.05 else 'ns'
        axes[i].set_title(f'{col}\nMW-U p={p_mw:.3f} {sig}')
    else:
        axes[i].set_title(col)

plt.tight_layout(); plt.show()
````

## 5 · Correlation Study — Pearson · Spearman · Kendall

````python
# Build encoded frame
df_enc = df[NUM_COLS + BOOL_COLS + [TARGET]].copy()
for c in BOOL_COLS: df_enc[c] = df_enc[c].astype(float)
df_enc['os_label_enc'] = LabelEncoder().fit_transform(df['os_label'].astype(str))
feat_cols = NUM_COLS + BOOL_COLS + ['os_label_enc']
df_corr = df_enc[feat_cols + [TARGET]].dropna()

corr_p = df_corr.corr(method='pearson')
corr_s = df_corr.corr(method='spearman')
corr_k = df_corr.corr(method='kendall')

# Fig 12 — Three correlation matrices
fig, axes = plt.subplots(1, 3, figsize=(26, 9))
fig.suptitle('Fig 12 — Correlation Matrices (Pearson / Spearman / Kendall)', fontsize=14, fontweight='bold')
mask = np.triu(np.ones_like(corr_p, dtype=bool))
kws  = dict(mask=mask, vmin=-1, vmax=1, center=0, cmap='RdYlGn',
            linewidths=0.3, annot=True, fmt='.2f', annot_kws={'size':6}, square=True)
sns.heatmap(corr_p, ax=axes[0], **kws); axes[0].set_title('Pearson')
sns.heatmap(corr_s, ax=axes[1], **kws); axes[1].set_title('Spearman')
sns.heatmap(corr_k, ax=axes[2], **kws); axes[2].set_title('Kendall')
plt.tight_layout(); plt.show()
````

````python
# Fig 13 — Correlation with target: all three metrics side-by-side
corr_with_target = pd.DataFrame({
    'Pearson':  [df_corr[c].corr(df_corr[TARGET], method='pearson')  for c in feat_cols],
    'Spearman': [df_corr[c].corr(df_corr[TARGET], method='spearman') for c in feat_cols],
    'Kendall':  [df_corr[c].corr(df_corr[TARGET], method='kendall')  for c in feat_cols],
}, index=feat_cols).sort_values('Spearman', ascending=False)

fig, ax = plt.subplots(figsize=(14, 7))
fig.suptitle('Fig 13 — Feature Correlation with duration_minutes', fontsize=14, fontweight='bold')
x_ = np.arange(len(corr_with_target)); w_ = 0.25
ax.bar(x_-w_,  corr_with_target['Pearson'],  w_, label='Pearson',  color='#3498db', alpha=0.85)
ax.bar(x_,     corr_with_target['Spearman'], w_, label='Spearman', color='#e67e22', alpha=0.85)
ax.bar(x_+w_,  corr_with_target['Kendall'],  w_, label='Kendall',  color='#2ecc71', alpha=0.85)
ax.axhline(0,   color='black', lw=0.8)
ax.axhline( 0.3, color='gray', ls=':', lw=1, label='|r|=0.3 threshold')
ax.axhline(-0.3, color='gray', ls=':', lw=1)
ax.set_xticks(x_)
ax.set_xticklabels(corr_with_target.index, rotation=45, ha='right', fontsize=9)
ax.set_ylabel('Correlation coefficient')
ax.legend(); ax.set_ylim(-1, 1)
plt.tight_layout(); plt.show()

print(corr_with_target.round(3).to_string())
````

````python
# Fig 14 — Spearman lollipop (diverging)
sp_sorted = corr_with_target['Spearman'].sort_values()

fig, ax = plt.subplots(figsize=(10, 8))
fig.suptitle('Fig 14 — Spearman Rank Correlation Lollipop', fontsize=14, fontweight='bold')
colors_ = ['#e74c3c' if v>0 else '#3498db' for v in sp_sorted.values]
ax.hlines(range(len(sp_sorted)), 0, sp_sorted.values, color=colors_, lw=2)
ax.scatter(sp_sorted.values, range(len(sp_sorted)), color=colors_, s=80, zorder=5)
ax.axvline(0, color='black', lw=0.8)
ax.axvline( 0.3, color='gray', ls='--', lw=1)
ax.axvline(-0.3, color='gray', ls='--', lw=1)
ax.set_yticks(range(len(sp_sorted)))
ax.set_yticklabels(sp_sorted.index)
ax.set_xlabel('Spearman rho with duration_minutes')
for i, (val, feat) in enumerate(zip(sp_sorted.values, sp_sorted.index)):
    ax.text(val+0.01*np.sign(val), i, f'{val:.3f}', va='center', fontsize=8)
plt.tight_layout(); plt.show()
````

````python
# Fig 15 — Correlation clustering dendrogram + reordered heatmap
corr_abs = df_corr[feat_cols].corr(method='spearman').abs()
dist_mat  = 1 - corr_abs
np.fill_diagonal(dist_mat.values, 0)
linkage   = hierarchy.ward(squareform(dist_mat.values))

fig, axes = plt.subplots(1, 2, figsize=(20, 8))
fig.suptitle('Fig 15 — Feature Clustering (Spearman Distance)', fontsize=14, fontweight='bold')

hierarchy.dendrogram(linkage, labels=feat_cols, leaf_rotation=90,
                     leaf_font_size=9, ax=axes[0], color_threshold=0.7*max(linkage[:,2]))
axes[0].set_title('Hierarchical Clustering Dendrogram')

order_idx = hierarchy.leaves_list(linkage)
ordered   = [feat_cols[i] for i in order_idx]
sns.heatmap(corr_abs[ordered].loc[ordered], cmap='RdYlGn', vmin=0, vmax=1,
            linewidths=0.3, annot=True, fmt='.2f', annot_kws={'size':6},
            ax=axes[1])
axes[1].set_title('Reordered Correlation Heatmap')
plt.tight_layout(); plt.show()
````

## 6 · Categorical & Boolean Deep Dive

## 8 · Outlier & Leverage Analysis

````python
# Fig 22 — IQR outlier detection
outlier_summary = {}
for col in NUM_COLS + [TARGET]:
    Q1, Q3 = df[col].quantile([0.25,0.75])
    IQR = Q3-Q1
    lo, hi = Q1-1.5*IQR, Q3+1.5*IQR
    outlier_summary[col] = {'Q1':Q1,'Q3':Q3,'IQR':IQR,'LF':lo,'UF':hi,
                            'Outliers':((df[col]<lo)|(df[col]>hi)).sum()}
out_df = pd.DataFrame(outlier_summary).T.round(3)

fig, axes = plt.subplots(2, 1, figsize=(14, 10))
fig.suptitle('Fig 22 — Outlier Detection', fontsize=14, fontweight='bold')

scale_cols = [c for c in NUM_COLS if df[c].std()>0]
df_scaled  = (df[scale_cols]-df[scale_cols].mean()) / df[scale_cols].std()
bp = axes[0].boxplot([df_scaled[c].dropna() for c in scale_cols], patch_artist=True,
                     medianprops={'color':'red','lw':2})
for patch, color in zip(bp['boxes'], PALETTE*10):
    patch.set_facecolor(color); patch.set_alpha(0.6)
axes[0].set_xticks(range(1, len(scale_cols)+1))
axes[0].set_xticklabels(scale_cols, rotation=45, ha='right', fontsize=8)
axes[0].set_title('Standardised Box Plots')
axes[0].set_ylabel('Z-score')

n_out = out_df['Outliers'].astype(int).sort_values(ascending=False)
axes[1].bar(n_out.index, n_out.values,
            color=['#e74c3c' if v>0 else '#95a5a6' for v in n_out.values], edgecolor='white')
axes[1].set_xticklabels(n_out.index, rotation=45, ha='right', fontsize=9)
axes[1].set_title('IQR Outlier Count per Feature')
plt.tight_layout(); plt.show()
print(out_df[['LF','UF','Outliers']].to_string())
````

````python
# Fig 23 — Cook's Distance + Residuals vs Fitted
X_ols = sm.add_constant(df_corr[feat_cols].copy())
y_ols = df_corr[TARGET]
model = sm.OLS(y_ols, X_ols).fit()
cooks = model.get_influence().cooks_distance[0]
thresh_cd = 4 / len(cooks)

fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle("Fig 23 — Cook's Distance & Residuals", fontsize=14, fontweight='bold')

axes[0].stem(range(len(cooks)), cooks, markerfmt='o', linefmt='grey', basefmt='black')
for i_, c_ in enumerate(cooks):
    if c_ > thresh_cd:
        axes[0].scatter(i_, c_, color='red', s=60, zorder=5)
        axes[0].text(i_+0.3, c_, f'row {i_}', fontsize=7, color='red')
axes[0].axhline(thresh_cd, color='red', ls='--', lw=1.5, label=f'4/n={thresh_cd:.4f}')
axes[0].set_xlabel('Row index'); axes[0].set_ylabel("Cook's D")
axes[0].set_title("Cook's Distance"); axes[0].legend()

sc_ = axes[1].scatter(model.fittedvalues, model.resid, c=cooks, cmap='YlOrRd', s=50, alpha=0.7)
plt.colorbar(sc_, ax=axes[1], label="Cook's D")
axes[1].axhline(0, color='black', lw=1)
axes[1].set_xlabel('Fitted'); axes[1].set_ylabel('Residuals')
axes[1].set_title('Residuals vs Fitted')
plt.tight_layout(); plt.show()
print(f'OLS R2={model.rsquared:.4f}, Adj R2={model.rsquared_adj:.4f}')
print(f'Influential rows (Cook>4/n): {(cooks>thresh_cd).sum()}')
````

## 10 · Statistical Feature Importance

````python
X_all = df_corr[feat_cols].copy()
y_all = df_corr[TARGET]

mi_scores = mutual_info_regression(X_all, y_all, random_state=SEED)
f_scores, _ = f_regression(X_all, y_all)

rf = RandomForestRegressor(n_estimators=500, random_state=SEED, n_jobs=-1)
rf.fit(X_all, y_all)
rf_imp = rf.feature_importances_

gb = GradientBoostingRegressor(n_estimators=300, random_state=SEED)
gb.fit(X_all, y_all)
gb_imp = gb.feature_importances_

importance_df = pd.DataFrame({
    'Mutual Info':  mi_scores / mi_scores.max(),
    'F-Score':      f_scores  / f_scores.max(),
    'RandomForest': rf_imp    / rf_imp.max(),
    'GradBoost':    gb_imp    / gb_imp.max(),
    'Spearman|r|':  corr_with_target['Spearman'].reindex(feat_cols).abs().values,
}, index=feat_cols)
importance_df['Composite'] = importance_df.mean(axis=1)
importance_df = importance_df.sort_values('Composite', ascending=False)
importance_df.round(4)
````

````python
# Fig 25 — Grouped bar + heatmap
methods = ['Mutual Info','F-Score','RandomForest','GradBoost','Spearman|r|']
fig, axes = plt.subplots(1, 2, figsize=(20, 8))
fig.suptitle('Fig 25 — Feature Importance: All Methods Compared', fontsize=14, fontweight='bold')

x_ = np.arange(len(importance_df)); w_ = 0.15
for j, method in enumerate(methods):
    axes[0].bar(x_+j*w_, importance_df[method], w_, label=method, color=PALETTE[j], alpha=0.85)
axes[0].set_xticks(x_+2*w_)
axes[0].set_xticklabels(importance_df.index, rotation=45, ha='right', fontsize=9)
axes[0].set_ylabel('Normalised Importance (0-1)')
axes[0].set_title('All Methods (normalised)')
axes[0].legend(fontsize=8)

sns.heatmap(importance_df[methods].T, cmap='YlOrRd', annot=True, fmt='.3f',
            annot_kws={'size':8}, linewidths=0.3, ax=axes[1],
            cbar_kws={'label':'Normalised importance'})
axes[1].set_title('Importance Heatmap')
plt.tight_layout(); plt.show()
````

## 11 · Dimensionality Reduction — PCA & t-SNE

````python
# Fig 28 — PCA: scree + biplot + 2D scatter
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_all)

pca_full = PCA(random_state=SEED).fit(X_scaled)
ev  = pca_full.explained_variance_ratio_
cum = np.cumsum(ev)
n85 = int(np.searchsorted(cum, 0.85)) + 1

pca2   = PCA(n_components=2, random_state=SEED)
X_pca  = pca2.fit_transform(X_scaled)

fig, axes = plt.subplots(1, 3, figsize=(21, 6))
fig.suptitle('Fig 28 — PCA Analysis', fontsize=14, fontweight='bold')

axes[0].bar(range(1,len(ev)+1), ev*100, color=PALETTE[0], alpha=0.8)
ax_r = axes[0].twinx()
ax_r.plot(range(1,len(cum)+1), cum*100, 'ro-', ms=5)
ax_r.axhline(85, color='gray', ls='--', lw=1)
ax_r.set_ylabel('Cumulative variance (%)')
axes[0].set_xlabel('PC'); axes[0].set_ylabel('Explained variance (%)')
axes[0].set_title(f'Scree Plot ({n85} PCs -> 85%)')

sc_ = axes[1].scatter(X_pca[:,0], X_pca[:,1], c=y_all.values, cmap='YlOrRd', s=60, alpha=0.8, edgecolors='white')
plt.colorbar(sc_, ax=axes[1], label='duration_minutes')
axes[1].set_xlabel(f'PC1 ({ev[0]*100:.1f}%)'); axes[1].set_ylabel(f'PC2 ({ev[1]*100:.1f}%)')
axes[1].set_title('PCA 2D: colour = duration')

loadings = pca2.components_.T
for i, feat in enumerate(feat_cols):
    axes[2].arrow(0,0,loadings[i,0]*3,loadings[i,1]*3,
                  head_width=0.08, head_length=0.05, fc=PALETTE[i%10], ec=PALETTE[i%10])
    axes[2].text(loadings[i,0]*3.3, loadings[i,1]*3.3, feat, fontsize=7,
                 ha='center', va='center', color=PALETTE[i%10])
axes[2].set_xlim(-4,4); axes[2].set_ylim(-4,4)
axes[2].axhline(0,color='gray',lw=0.5); axes[2].axvline(0,color='gray',lw=0.5)
axes[2].set_title('PCA Biplot (feature loadings)')
plt.tight_layout(); plt.show()
````

## 12 · Feature Engineering Candidates

````python
# Fig 30 — Engineered features vs original
df_eng = df.copy()
df_eng['steps_per_yaml_line'] = df_eng['total_steps']         / (df_eng['yaml_line_count']+1)
df_eng['actions_per_step']    = df_eng['unique_actions_used'] / (df_eng['total_steps']+1)
df_eng['complexity_per_job']  = df_eng['code_complexity']     / (df_eng['job_count']+1)
df_eng['log_yaml_line_count'] = np.log1p(df_eng['yaml_line_count'])
df_eng['log_total_steps']     = np.log1p(df_eng['total_steps'])
df_eng['total_steps_sq']      = df_eng['total_steps']**2
df_eng['yaml_x_jobs']         = df_eng['yaml_line_count'] * df_eng['job_count']
df_eng['matrix_x_jobs']       = df_eng['matrix_permutations'] * df_eng['job_count']
df_eng['n_ci_features']       = (df_eng['is_using_setup_actions'].astype(int) +
                                  df_eng['is_using_cache'].astype(int) +
                                  df_eng['is_using_docker_actions'].astype(int) +
                                  df_eng['uses_matrix_strategy'].astype(int))

eng_cols = ['steps_per_yaml_line','actions_per_step','complexity_per_job',
            'log_yaml_line_count','log_total_steps','total_steps_sq',
            'yaml_x_jobs','matrix_x_jobs','n_ci_features']

eng_corrs = pd.DataFrame({
    'Pearson':  [df_eng[c].corr(df_eng[TARGET]) for c in eng_cols],
    'Spearman': [df_eng[c].corr(df_eng[TARGET], method='spearman') for c in eng_cols],
}, index=eng_cols).sort_values('Spearman', ascending=False)

orig_top = corr_with_target[['Pearson','Spearman']].head(8)
combined = pd.concat([orig_top, eng_corrs], keys=['Original','Engineered'])

fig, ax = plt.subplots(figsize=(14, 8))
fig.suptitle('Fig 30 — Engineered vs Original Features: Correlation with Target', fontsize=13, fontweight='bold')
x_ = np.arange(len(combined)); w_ = 0.35
ax.bar(x_-w_/2, combined['Pearson'],  w_, label='Pearson',  color='#3498db', alpha=0.8)
ax.bar(x_+w_/2, combined['Spearman'], w_, label='Spearman', color='#e67e22', alpha=0.8)
ax.set_xticks(x_)
ax.set_xticklabels([f'[{lvl}] {feat}' for lvl,feat in combined.index], rotation=50, ha='right', fontsize=8)
ax.axhline(0, color='black', lw=0.7)
ax.axhline(0.3, color='gray', ls=':', lw=1)
ax.set_ylabel('Correlation with duration_minutes')
ax.legend()
plt.tight_layout(); plt.show()
print(eng_corrs.round(4).to_string())
````

## 13 · Final Feature Ranking & ML Model Recommendation
