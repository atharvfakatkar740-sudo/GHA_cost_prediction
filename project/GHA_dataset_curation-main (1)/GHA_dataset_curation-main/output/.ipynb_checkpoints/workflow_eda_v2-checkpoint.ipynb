# GitHub Actions Workflow Duration — EDA
### Feature Analysis for Cold-Start Duration Prediction

**Goal:** Identify features with predictive signal for `duration_minutes`, understand dataset structure, and specify the optimal feature set for ML model training.

> `total_cost_usd` is excluded throughout — it is a deterministic function of the target and constitutes data leakage.

````python
import warnings; warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from scipy import stats
from scipy.stats import pearsonr, spearmanr, kendalltau, kruskal, shapiro
from sklearn.feature_selection import mutual_info_regression
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from statsmodels.stats.outliers_influence import variance_inflation_factor

# ── Style ──────────────────────────────────────────────────────────────────
BG       = '#0f1117'
PANEL    = '#1a1d2e'
GRID     = '#2a2d4a'
TICK     = '#a0a8cc'
TEXT     = '#e0e4ff'
C1, C2, C3, C4, C5, C6 = '#7c83e8','#f7c59f','#57e0a6','#e8607c','#c57ce8','#60c9e8'
PALETTE  = [C1, C2, C3, C4, C5, C6, '#e8e060']

plt.rcParams.update({
    'figure.facecolor': BG, 'axes.facecolor': PANEL, 'axes.edgecolor': GRID,
    'axes.labelcolor': TEXT, 'axes.titlecolor': TEXT, 'axes.grid': True,
    'grid.color': GRID, 'grid.linewidth': 0.5, 'xtick.color': TICK,
    'ytick.color': TICK, 'text.color': TEXT, 'font.family': 'DejaVu Sans',
    'font.size': 11, 'axes.titlesize': 12, 'axes.titleweight': 'bold',
    'figure.titlesize': 14, 'figure.titleweight': 'bold',
})
sns.set_palette(PALETTE)
TARGET = 'duration_minutes'
print('Setup complete ✓')
````

---
## 1 · Load, Type-Cast & Outlier Removal

````python
df_raw = pd.read_csv('workflow_features.csv')

# Cast booleans
bool_cols = ['uses_matrix_strategy','fail_fast','is_using_setup_actions',
             'is_using_docker_actions','is_using_cache']
for c in bool_cols:
    df_raw[c] = df_raw[c].map({'True': 1, 'False': 0, True: 1, False: 0}).fillna(0).astype(int)
for c in bool_cols:
    df_raw[c] = df_raw[c].map({'True': 1, 'False': 0, True: 1, False: 0}).astype(int)

df_raw['os_is_ubuntu'] = (df_raw['os_label'] == 'ubuntu').astype(int)

# ── Outlier removal on TARGET using IQR ───────────────────────────────────
Q1, Q3 = df_raw[TARGET].quantile([0.25, 0.75])
IQR    = Q3 - Q1
lo, hi = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR

outliers = df_raw[~df_raw[TARGET].between(lo, hi)]
df       = df_raw[df_raw[TARGET].between(lo, hi)].copy().reset_index(drop=True)

print(f'Raw rows   : {len(df_raw):>4}')
print(f'Outliers   : {len(outliers):>4}  (duration > {hi:.2f} min)')
print(f'Clean rows : {len(df):>4}')
print(f'\nOutlier durations  : {sorted(outliers[TARGET].unique())}')
print(f'Outlier workflows  : {sorted(outliers["workflow_name"].unique())}')
print(f'\nClean duration range: {df[TARGET].min():.3f} – {df[TARGET].max():.3f} min')
````

---
## 2 · Target Distribution: Before vs After Outlier Removal

````python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Target Variable — duration_minutes')

# Before
ax = axes[0]
ax.hist(df_raw[TARGET], bins=25, color=C4, edgecolor=BG, alpha=0.85)
ax.axvline(hi, color=C2, lw=2, ls='--', label=f'IQR upper fence ({hi:.1f})')
ax.set_xlabel('duration_minutes'); ax.set_ylabel('Count')
ax.set_title('Full Dataset (with outliers)')
ax.legend(fontsize=9)

# After — histogram + KDE
ax = axes[1]
ax.hist(df[TARGET], bins=20, color=C1, edgecolor=BG, alpha=0.85, density=True)
kde_x = np.linspace(df[TARGET].min()-0.1, df[TARGET].max()+0.1, 300)
kde   = stats.gaussian_kde(df[TARGET], bw_method=0.5)
ax.plot(kde_x, kde(kde_x), color=C2, lw=2.5, label='KDE')
ax.axvline(df[TARGET].mean(),   color=C3, lw=1.8, ls='--', label=f'Mean  {df[TARGET].mean():.2f}')
ax.axvline(df[TARGET].median(), color=C4, lw=1.8, ls=':',  label=f'Median {df[TARGET].median():.2f}')
ax.set_xlabel('duration_minutes'); ax.set_ylabel('Density')
ax.set_title('Clean Dataset — Histogram + KDE')
ax.legend(fontsize=9)

# After — box+violin
ax = axes[2]
parts = ax.violinplot(df[TARGET], positions=[0], showmedians=True)
for pc in parts['bodies']:
    pc.set_facecolor(C1); pc.set_alpha(0.5)
parts['cmedians'].set_color(C2)
parts['cmins'].set_color(TICK); parts['cmaxes'].set_color(TICK)
parts['cbars'].set_color(TICK)
bp = ax.boxplot(df[TARGET], positions=[0], widths=0.12, patch_artist=True,
                boxprops={'facecolor': C4, 'alpha': 0.7},
                medianprops={'color': 'white', 'lw': 2},
                whiskerprops={'color': TICK}, capprops={'color': TICK},
                flierprops={'marker': 'o', 'color': C4, 'ms': 5})
ax.set_xticks([]); ax.set_ylabel('duration_minutes')
ax.set_title('Clean Dataset — Violin + Box')

plt.tight_layout(); plt.show()

stat_w, p_w = shapiro(df[TARGET])
print(f'Shapiro-Wilk p = {p_w:.4f} → {"NOT normal" if p_w < 0.05 else "Normal"}')
print(f'Skewness = {df[TARGET].skew():.3f}  |  Kurtosis = {df[TARGET].kurtosis():.3f}')
````

---
## 3 · Dataset Structure Audit: Feature Variance

Before computing correlations, we must identify which features actually vary in the clean dataset.
A feature with zero variance carries no predictive signal and must be dropped before any model is trained.

````python
candidate_features = [
    'yaml_line_count','yaml_depth','total_steps','avg_steps_per_job',
    'unique_actions_used','matrix_dimensions','matrix_permutations',
    'timeout_minutes','env_var_count','if_condition_count',
    'needs_dependencies_count','code_complexity','job_count',
    'uses_matrix_strategy','fail_fast','is_using_setup_actions',
    'is_using_docker_actions','is_using_cache','os_is_ubuntu',
]

variance_report = []
for col in candidate_features:
    s = df[col].std()
    u = df[col].nunique()
    variance_report.append({'Feature': col, 'Std Dev': round(s,4),
                            'Unique Values': u, 'Has Variance': s > 0})

var_df = pd.DataFrame(variance_report)
zero_var = var_df[~var_df['Has Variance']]['Feature'].tolist()
has_var  = var_df[ var_df['Has Variance']]['Feature'].tolist()

# ── Plot ──────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(18, 5))
fig.suptitle('Feature Variance Audit (clean dataset)')

# Std dev bar chart
ax = axes[0]
sorted_var = var_df.sort_values('Std Dev', ascending=True)
bar_colors = [C3 if row['Has Variance'] else C4 for _, row in sorted_var.iterrows()]
ax.barh(sorted_var['Feature'], sorted_var['Std Dev'],
        color=bar_colors, edgecolor='none')
ax.axvline(0, color='white', lw=0.5)
ax.set_xlabel('Standard Deviation')
ax.set_title('Feature Std Dev — Green=has variance, Red=constant (zero-variance)')
from matplotlib.patches import Patch
ax.legend(handles=[Patch(color=C3, label='Has variance (usable)'),
                   Patch(color=C4, label='Constant — DROP')], fontsize=9)

# Unique value count
ax = axes[1]
sorted_uniq = var_df.sort_values('Unique Values', ascending=True)
bar_colors2 = [C3 if row['Has Variance'] else C4 for _, row in sorted_uniq.iterrows()]
ax.barh(sorted_uniq['Feature'], sorted_uniq['Unique Values'],
        color=bar_colors2, edgecolor='none')
ax.set_xlabel('Number of Unique Values')
ax.set_title('Unique Value Count per Feature')

plt.tight_layout(); plt.show()

print(f'ZERO-VARIANCE features (drop before training): {zero_var}')
print(f'\nUSABLE features ({len(has_var)} total): {has_var}')
````

---
## 4 · Duration by Workflow Type — The Primary Signal

All numeric YAML features (steps, lines, actions) are extracted from the workflow YAML file, which is fixed per workflow. Since this dataset covers a **single repository**, workflows of the same type share identical YAML-derived feature values. Consequently, **`workflow_name` (the categorical workflow type) is the dominant signal** — it encodes the combined effect of all YAML properties.

````python
wf_order = df.groupby('workflow_name')[TARGET].mean().sort_values(ascending=False).index
wf_short = [w[:28] + ('…' if len(w) > 28 else '') for w in wf_order]

fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('Duration by Workflow Type — Primary Categorical Signal')

# Box per workflow
ax = axes[0]
data = [df[df['workflow_name'] == w][TARGET].values for w in wf_order]
bp = ax.boxplot(data, vert=True, patch_artist=True,
                medianprops={'color': 'white', 'lw': 2.5},
                whiskerprops={'color': TICK}, capprops={'color': TICK},
                flierprops={'marker': 'o', 'color': C4, 'ms': 5})
for i, patch in enumerate(bp['boxes']):
    patch.set_facecolor(PALETTE[i % len(PALETTE)]); patch.set_alpha(0.8)
ax.set_xticks(range(1, len(wf_order)+1))
ax.set_xticklabels(wf_short, rotation=35, ha='right', fontsize=8)
ax.set_ylabel('duration_minutes')
ax.set_title('Duration Distribution by Workflow')

# Strip + mean dot
ax = axes[1]
for i, w in enumerate(wf_order):
    sub = df[df['workflow_name'] == w][TARGET].values
    jitter = np.random.uniform(-0.18, 0.18, len(sub))
    ax.scatter(np.full(len(sub), i) + jitter, sub,
               color=PALETTE[i % len(PALETTE)], s=28, alpha=0.75, zorder=3)
    ax.scatter(i, sub.mean(), color='white', s=90, zorder=5, marker='D')
ax.set_xticks(range(len(wf_order)))
ax.set_xticklabels(wf_short, rotation=35, ha='right', fontsize=8)
ax.set_ylabel('duration_minutes')
ax.set_title('Strip Plot — White ◆ = Mean')

# Mean + 95% CI bar
ax = axes[2]
means, cis, ns = [], [], []
for w in wf_order:
    sub = df[df['workflow_name'] == w][TARGET].values
    m = sub.mean()
    ci = 1.96 * sub.std() / np.sqrt(len(sub)) if len(sub) > 1 else 0
    means.append(m); cis.append(ci); ns.append(len(sub))
ax.bar(range(len(wf_order)), means, color=PALETTE[:len(wf_order)],
       edgecolor=BG, alpha=0.85, yerr=cis,
       error_kw={'ecolor': TICK, 'lw': 1.5, 'capsize': 5})
ax.set_xticks(range(len(wf_order)))
ax.set_xticklabels(wf_short, rotation=35, ha='right', fontsize=8)
ax.set_ylabel('Mean duration_minutes')
ax.set_title('Mean ± 95% CI per Workflow')
for i, (m, n) in enumerate(zip(means, ns)):
    ax.text(i, m + cis[i] + 0.02, f'n={n}', ha='center', fontsize=8, color=TICK)

plt.tight_layout(); plt.show()

# Kruskal-Wallis test
stat_kw, p_kw = kruskal(*data)
print(f'Kruskal-Wallis test: H={stat_kw:.3f}, p={p_kw:.6f}')
print(f'→ {"Significant" if p_kw < 0.05 else "Not significant"} difference across workflow types (α=0.05)')
````

## 5 · YAML Features per Workflow — Encoding Analysis

Since all numeric YAML features are constant within this single-repo dataset, we visualise how they map to workflow types. These features **do carry between-workflow signal** and will be valid predictors when the dataset is expanded to multiple repositories.

````python
# Aggregate YAML features and mean duration per workflow type
yaml_features = ['total_steps','unique_actions_used','yaml_line_count',
                 'env_var_count','avg_steps_per_job']

wf_agg = df.groupby('workflow_name').agg(
    mean_duration=(TARGET, 'mean'),
    std_duration=(TARGET, 'std'),
    **{f: (f, 'first') for f in yaml_features},
    is_using_docker_actions=('is_using_docker_actions', 'first'),
    is_using_cache=('is_using_cache', 'first'),
    is_using_setup_actions=('is_using_setup_actions', 'first'),
    os_is_ubuntu=('os_is_ubuntu', 'mean'),
).reset_index()
wf_agg = wf_agg.sort_values('mean_duration', ascending=False)
wf_agg['wf_short'] = wf_agg['workflow_name'].str[:30]

fig, axes = plt.subplots(2, 3, figsize=(20, 10))
fig.suptitle('YAML Features vs Mean Duration per Workflow Type')
axes = axes.flatten()

plot_feats = [
    ('total_steps', 'Total Steps'),
    ('unique_actions_used', 'Unique Actions Used'),
    ('yaml_line_count', 'YAML Line Count'),
    ('env_var_count', 'Env Var Count'),
    ('avg_steps_per_job', 'Avg Steps per Job'),
    ('is_using_docker_actions', 'Uses Docker Actions'),
]

colors_wf = {w: PALETTE[i % len(PALETTE)] for i, w in enumerate(wf_agg['workflow_name'])}

for ax, (feat, label) in zip(axes, plot_feats):
    x = wf_agg[feat].values.astype(float)
    y = wf_agg['mean_duration'].values
    dot_colors = [colors_wf[w] for w in wf_agg['workflow_name']]

    # Scatter sized by n
    ax.scatter(x + np.random.uniform(-0.05, 0.05, len(x)), y,
               s=120, c=dot_colors, edgecolors='white', lw=0.8, zorder=4)

    # Label each point
    for xi, yi, wn in zip(x, y, wf_agg['wf_short']):
        ax.annotate(wn, (xi, yi), textcoords='offset points', xytext=(5, 3),
                    fontsize=7, color=TICK)

    # Regression line if variance exists
    if np.std(x) > 0:
        m, b, r, p, _ = stats.linregress(x, y)
        xs = np.linspace(x.min(), x.max(), 80)
        ax.plot(xs, m*xs + b, color='white', lw=1.5, ls='--', alpha=0.5,
                label=f'r={r:.2f}, p={p:.3f}')
        ax.legend(fontsize=8)

    ax.set_xlabel(label)
    ax.set_ylabel('Mean duration_minutes')
    ax.set_title(f'{label} vs Duration')

plt.tight_layout(); plt.show()

print('\nWorkflow-level feature table (each row = one workflow type):')
display(wf_agg[['wf_short','mean_duration','std_duration'] + yaml_features +
               ['is_using_docker_actions','is_using_cache']]
        .round(3).set_index('wf_short'))
````

## 6 · Correlation Metrics — All Features vs `duration_minutes`

We compute Pearson, Spearman, Kendall τ, and Mutual Information for every feature with non-zero variance. Encoding is applied where needed: `os_label` → binary, `workflow_name` → ordinal (sorted by mean duration, so the encoding preserves the natural ordering).

````python
# Encode workflow_name by mean duration rank (preserves natural ordering)
wf_rank = df.groupby('workflow_name')[TARGET].mean().rank().to_dict()
df['workflow_name_rank'] = df['workflow_name'].map(wf_rank)

# Final usable feature set (only non-zero-variance features)
analysis_features = [
    'workflow_name_rank',    # ordinal-encoded by mean duration
    'os_is_ubuntu',          # binary
    'is_using_docker_actions',
    'is_using_cache',
    'is_using_setup_actions',
    'total_steps',
    'unique_actions_used',
    'yaml_line_count',
    'env_var_count',
    'avg_steps_per_job',
]

y = df[TARGET].values
corr_results = []

for feat in analysis_features:
    x = df[feat].values.astype(float)
    if x.std() == 0:
        continue
    p_r,  p_p = pearsonr(x, y)
    s_r,  s_p = spearmanr(x, y)
    k_r,  k_p = kendalltau(x, y)
    mi         = mutual_info_regression(
        x.reshape(-1,1), y, random_state=42)[0]
    corr_results.append({
        'Feature':      feat,
        'Pearson r':    round(p_r, 4),
        'Pearson p':    round(p_p, 4),
        'Spearman ρ':   round(s_r, 4),
        'Spearman p':   round(s_p, 4),
        'Kendall τ':    round(k_r, 4),
        'Kendall p':    round(k_p, 4),
        'Mutual Info':  round(mi, 5),
    })

corr_df = pd.DataFrame(corr_results)
corr_df['|Pearson|']  = corr_df['Pearson r'].abs()
corr_df['|Spearman|'] = corr_df['Spearman ρ'].abs()
corr_df['|Kendall|']  = corr_df['Kendall τ'].abs()
corr_df = corr_df.sort_values('|Spearman|', ascending=False).reset_index(drop=True)

print('Correlation table (sorted by |Spearman|):')
display(corr_df[['Feature','Pearson r','Pearson p','Spearman ρ','Spearman p',
                  'Kendall τ','Kendall p','Mutual Info']]
        .style.background_gradient(cmap='RdYlGn', subset=['Pearson r','Spearman ρ','Kendall τ'])
        .background_gradient(cmap='Blues', subset=['Mutual Info'])
        .format({'Pearson r':'{:.4f}','Pearson p':'{:.4f}',
                 'Spearman ρ':'{:.4f}','Spearman p':'{:.4f}',
                 'Kendall τ':'{:.4f}','Kendall p':'{:.4f}',
                 'Mutual Info':'{:.5f}'}))
````

## 7 · Correlation Metrics — Side-by-Side Visualisation

````python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
fig.suptitle('All Correlation Metrics vs duration_minutes')

feat_labels = corr_df['Feature'].tolist()
x_idx = np.arange(len(feat_labels))
w = 0.22

# ── Grouped bar: Pearson / Spearman / Kendall ─────────────────────────────
ax = axes[0]
ax.bar(x_idx - w, corr_df['Pearson r'],  w, label='Pearson r',  color=C1, alpha=0.9, edgecolor=BG)
ax.bar(x_idx,     corr_df['Spearman ρ'], w, label='Spearman ρ', color=C2, alpha=0.9, edgecolor=BG)
ax.bar(x_idx + w, corr_df['Kendall τ'],  w, label='Kendall τ',  color=C3, alpha=0.9, edgecolor=BG)
ax.axhline(0, color='white', lw=0.8)
ax.set_xticks(x_idx)
ax.set_xticklabels(feat_labels, rotation=40, ha='right', fontsize=9)
ax.set_ylabel('Correlation Coefficient')
ax.set_title('Pearson / Spearman / Kendall\n(sorted by |Spearman|)')
ax.legend(fontsize=9)
# Significance stars above bars
for i, (_, row) in enumerate(corr_df.iterrows()):
    p_val = row['Spearman p']
    star = '***' if p_val < 0.001 else '**' if p_val < 0.01 else '*' if p_val < 0.05 else ''
    if star:
        ypos = max(row['Pearson r'], row['Spearman ρ'], row['Kendall τ']) + 0.02
        ax.text(i, ypos, star, ha='center', fontsize=10, color=C2)

# ── Mutual Information bar ────────────────────────────────────────────────
ax = axes[1]
mi_sorted = corr_df.sort_values('Mutual Info', ascending=True)
bar_colors = [C1 if v > mi_sorted['Mutual Info'].quantile(0.6) else '#3a3d6a'
              for v in mi_sorted['Mutual Info']]
ax.barh(mi_sorted['Feature'], mi_sorted['Mutual Info'],
        color=bar_colors, edgecolor='none')
ax.set_xlabel('Mutual Information Score')
ax.set_title('Mutual Information\n(captures non-linear relationships)')
for i, (_, row) in enumerate(mi_sorted.iterrows()):
    ax.text(row['Mutual Info'] + 0.001, i, f'{row["Mutual Info"]:.5f}',
            va='center', fontsize=8.5, color=TICK)

plt.tight_layout(); plt.show()
````

## 8 · Correlation Agreement — Pearson vs Spearman Divergence

When Spearman ρ >> Pearson r for a feature, the relationship is **monotonic but non-linear** — tree-based models will capture this better than linear regression.

````python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Correlation Agreement & Non-Linearity Diagnostic')

# Pearson vs Spearman scatter
ax = axes[0]
# Reuses colors from the palette until it matches the dataframe length
point_colors = [PALETTE[i % len(PALETTE)] for i in range(len(corr_df))]

ax.scatter(corr_df['|Pearson|'], corr_df['|Spearman|'], 
           s=100, c=point_colors, edgecolors='white', lw=0.6, zorder=4)
for _, row in corr_df.iterrows():
    ax.annotate(row['Feature'], (row['|Pearson|'], row['|Spearman|']),
                textcoords='offset points', xytext=(5, 4), fontsize=8.5, color=TICK)
lim = max(corr_df[['|Pearson|','|Spearman|']].max()) + 0.08
ax.plot([0, lim], [0, lim], 'w--', lw=1, alpha=0.4, label='y = x (perfect agreement)')
ax.set_xlim(0, lim); ax.set_ylim(0, lim)
ax.set_xlabel('|Pearson r|')
ax.set_ylabel('|Spearman ρ|')
ax.set_title('Points ABOVE the line → non-linear relationship')
ax.legend(fontsize=9)

# Divergence bar (|Spearman| - |Pearson|)
ax = axes[1]
corr_df['Divergence'] = corr_df['|Spearman|'] - corr_df['|Pearson|']
div_sorted = corr_df.sort_values('Divergence', ascending=True)
div_colors = [C3 if v > 0.05 else C4 if v < -0.05 else C2
              for v in div_sorted['Divergence']]
ax.barh(div_sorted['Feature'], div_sorted['Divergence'],
        color=div_colors, edgecolor='none')
ax.axvline(0, color='white', lw=1)
ax.axvline( 0.05, color=TICK, lw=1, ls=':', alpha=0.6)
ax.axvline(-0.05, color=TICK, lw=1, ls=':', alpha=0.6)
ax.set_xlabel('|Spearman| − |Pearson| (divergence)')
ax.set_title('Non-Linearity Indicator\nGreen > 0 → non-linear | Red < 0 → linear dominates')

plt.tight_layout(); plt.show()
````

## 9 · Feature vs Duration: Scatter Plots (Top Features)

````python
# Top 6 features by |Spearman|
top6 = corr_df.nlargest(6, '|Spearman|')['Feature'].tolist()

fig, axes = plt.subplots(2, 3, figsize=(19, 10))
fig.suptitle('Top 6 Features vs duration_minutes — Scatter with LOWESS Trend')
axes = axes.flatten()

from statsmodels.nonparametric.smoothers_lowess import lowess

# Colour points by workflow type for context
wf_to_color = {w: PALETTE[i % len(PALETTE)] for i, w in enumerate(df['workflow_name'].unique())}
dot_colors_all = [wf_to_color[w] for w in df['workflow_name']]

for i, feat in enumerate(top6):
    ax = axes[i]
    x = df[feat].values.astype(float)
    y = df[TARGET].values

    ax.scatter(x + np.random.uniform(-0.05, 0.05, len(x)), y,
               c=dot_colors_all, s=32, alpha=0.8, edgecolors='white', lw=0.3, zorder=3)

    # OLS line
    if x.std() > 0:
        m, b, r, p_val, _ = stats.linregress(x, y)
        xs = np.linspace(x.min(), x.max(), 100)
        ax.plot(xs, m*xs + b, color='white', lw=1.8, ls='--',
                label=f'OLS r={r:.3f}', alpha=0.7)

        # LOWESS
        lo_out = lowess(y, x, frac=0.6, return_sorted=True)
        ax.plot(lo_out[:, 0], lo_out[:, 1], color=C2, lw=2.2, label='LOWESS')

        p_star = '***' if p_val < 0.001 else '**' if p_val < 0.01 else '*' if p_val < 0.05 else 'ns'
        ax.text(0.97, 0.97, p_star, transform=ax.transAxes,
                ha='right', va='top', fontsize=12, color=C2)
        ax.legend(fontsize=8)

    ax.set_xlabel(feat, fontsize=10)
    ax.set_ylabel('duration_minutes')
    ax.set_title(feat)

# Workflow-colour legend (shared)
from matplotlib.lines import Line2D
legend_handles = [Line2D([0],[0], marker='o', color='w',
                          markerfacecolor=c, markersize=8, label=w[:25])
                  for w, c in wf_to_color.items()]
fig.legend(handles=legend_handles, loc='lower center', ncol=3,
           fontsize=8, framealpha=0.3, bbox_to_anchor=(0.5, -0.04))

plt.tight_layout()
plt.subplots_adjust(bottom=0.13)
plt.show()
````

## 10 · Categorical Features vs Duration

````python
cat_feats = ['os_is_ubuntu','is_using_docker_actions','is_using_cache','is_using_setup_actions']
cat_labels = ['OS = Ubuntu', 'Uses Docker', 'Uses Cache', 'Uses Setup Actions']

fig, axes = plt.subplots(1, 4, figsize=(20, 6))
fig.suptitle('Binary Feature Groups vs duration_minutes — Violin + Strip + KW test')

for ax, feat, label in zip(axes, cat_feats, cat_labels):
    groups   = sorted(df[feat].unique())
    grp_data = [df[df[feat] == g][TARGET].values for g in groups]
    grp_means = [g.mean() for g in grp_data]

    if all(len(g) > 1 for g in grp_data):
        parts = ax.violinplot(grp_data, positions=range(len(groups)), showmedians=True)
        for j, pc in enumerate(parts['bodies']):
            pc.set_facecolor(PALETTE[j]); pc.set_alpha(0.4)
        parts['cmedians'].set_color('white')
        for key in ('cmins','cmaxes','cbars'):
            parts[key].set_color(TICK)

    for j, (grp, color) in enumerate(zip(grp_data, PALETTE)):
        jitter = np.random.uniform(-0.1, 0.1, len(grp))
        ax.scatter(np.full(len(grp), j) + jitter, grp,
                   color=color, s=28, alpha=0.9, zorder=4)
        ax.scatter(j, grp.mean(), color='white', s=80, zorder=5, marker='D')

    ax.set_xticks(range(len(groups)))
    ax.set_xticklabels(['False (0)' if g == 0 else 'True (1)' for g in groups])
    ax.set_ylabel('duration_minutes')
    ax.set_title(label)

    # Effect size (Cohen's d)
    if len(grp_data) == 2 and all(len(g) > 1 for g in grp_data):
        stat_kw, p_kw = kruskal(*grp_data)
        pooled_std = np.sqrt((grp_data[0].var() + grp_data[1].var()) / 2)
        d = abs(grp_data[0].mean() - grp_data[1].mean()) / pooled_std if pooled_std > 0 else 0
        sig = '***' if p_kw < 0.001 else '**' if p_kw < 0.01 else '*' if p_kw < 0.05 else 'ns'
        ax.text(0.97, 0.97, f'KW {sig}\nd={d:.2f}',
                transform=ax.transAxes, ha='right', va='top',
                fontsize=9, color=TICK)

plt.tight_layout(); plt.show()
````

## 12 · Multicollinearity — Variance Inflation Factor

````python
vif_cols = [c for c in analysis_features if df[c].std() > 0]
X_vif    = df[vif_cols].astype(float)

vif_vals = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]
vif_df   = pd.DataFrame({'Feature': vif_cols, 'VIF': vif_vals}).sort_values('VIF', ascending=False)

fig, ax = plt.subplots(figsize=(10, 5))
fig.suptitle('Variance Inflation Factor — Multicollinearity Check')

vif_colors = ['#e87c60' if v > 10 else C2 if v > 5 else C3 for v in vif_df['VIF']]
ax.barh(vif_df['Feature'][::-1], vif_df['VIF'][::-1], color=vif_colors[::-1], edgecolor='none')
ax.axvline(5,  color=C2,    lw=1.8, ls='--', label='VIF = 5  (moderate threshold)')
ax.axvline(10, color='#e87c60', lw=1.8, ls='--', label='VIF = 10 (high — consider dropping)')
ax.set_xlabel('VIF Score')
ax.set_title('VIF per Feature  (green < 5 = safe, orange 5–10 = moderate, red > 10 = problematic)')
ax.legend(fontsize=9)
for i, (_, row) in enumerate(vif_df[::-1].iterrows()):
    ax.text(row['VIF'] + 0.05, i, f'{row["VIF"]:.1f}', va='center', fontsize=9, color=TICK)

plt.tight_layout(); plt.show()

high_vif = vif_df[vif_df['VIF'] > 10]
if len(high_vif):
    print(f'⚠  Features with VIF > 10 (collinear, consider dropping for linear models):')
    for _, r in high_vif.iterrows():
        print(f'   {r["Feature"]}: VIF = {r["VIF"]:.1f}')
else:
    print('All features have VIF ≤ 10  ✓')
````

## 13 · Random Forest Feature Importance

Impurity-based importance from a Random Forest — a model-based (non-linear) importance measure, used here as an additional ranking perspective alongside correlation metrics.

````python
rf_feats = [c for c in analysis_features if df[c].std() > 0]
X_rf = df[rf_feats].astype(float).values
y_rf = df[TARGET].values

np.random.seed(42)
rf = RandomForestRegressor(n_estimators=200, max_depth=6, min_samples_leaf=2,
                            random_state=42, n_jobs=-1)
rf.fit(X_rf, y_rf)

rf_imp = pd.Series(rf.feature_importances_, index=rf_feats).sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(10, 5))
fig.suptitle('Random Forest Feature Importance (Impurity-based)')

bar_colors = [C1 if i < 3 else C2 if i < 6 else '#444466' for i in range(len(rf_imp))]
ax.bar(range(len(rf_imp)), rf_imp.values, color=bar_colors, edgecolor=BG, alpha=0.9)
ax.set_xticks(range(len(rf_imp)))
ax.set_xticklabels(rf_imp.index, rotation=35, ha='right', fontsize=9)
ax.set_ylabel('Mean Decrease in Impurity')
ax.set_title(f'In-sample R² = {rf.score(X_rf, y_rf):.4f}')
for i, v in enumerate(rf_imp.values):
    ax.text(i, v + 0.002, f'{v:.3f}', ha='center', fontsize=8.5, color=TICK)

plt.tight_layout(); plt.show()
````

## 14 · Final Feature Ranking & ML Recommendation

````python
# ── Unified ranking: normalise all 4 metrics to [0,1] rank score ──────────
def rank_norm(series):
    r = series.rank(ascending=True)
    return (r / len(r)).round(4)

rank_df = corr_df[corr_df['Feature'].isin(rf_feats)].copy()
rank_df['RF Importance'] = rank_df['Feature'].map(rf_imp.to_dict()).fillna(0)

for col in ['|Pearson|','|Spearman|','|Kendall|','Mutual Info','RF Importance']:
    rank_df[f'R_{col}'] = rank_norm(rank_df[col])

r_cols = [c for c in rank_df.columns if c.startswith('R_')]
rank_df['Composite'] = rank_df[r_cols].mean(axis=1)
rank_df = rank_df.sort_values('Composite', ascending=False).reset_index(drop=True)
rank_df.index += 1

# ── Plot ──────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
fig.suptitle('Final Feature Ranking — Composite Score')

# Stacked rank-metric breakdown
ax = axes[0]
rank_metric_names = ['R_|Pearson|','R_|Spearman|','R_|Kendall|','R_Mutual Info','R_RF Importance']
bar_w = 0.6
bottoms = np.zeros(len(rank_df))
for j, (metric, color) in enumerate(zip(rank_metric_names, PALETTE)):
    vals = rank_df[metric].values
    ax.bar(range(len(rank_df)), vals / 5, bottom=bottoms, width=bar_w,
           color=color, edgecolor=BG, alpha=0.85,
           label=metric.replace('R_','').replace('|',''))
    bottoms += vals / 5
ax.set_xticks(range(len(rank_df)))
ax.set_xticklabels(rank_df['Feature'], rotation=40, ha='right', fontsize=9)
ax.set_ylabel('Stacked normalised rank (each metric = 0.2 max)')
ax.set_title('Metric Contribution per Feature')
ax.legend(fontsize=8, loc='upper right')

# Composite score bar with tier zones
ax = axes[1]
tier_colors = [C1 if i < 2 else C2 if i < 5 else '#555577' for i in range(len(rank_df))]
ax.bar(range(len(rank_df)), rank_df['Composite'],
       color=tier_colors, edgecolor=BG, alpha=0.9)
ax.set_xticks(range(len(rank_df)))
ax.set_xticklabels(rank_df['Feature'], rotation=40, ha='right', fontsize=9)
ax.set_ylabel('Composite Rank Score (0–1)')
ax.set_title('Overall Feature Importance — Composite')
for i, v in enumerate(rank_df['Composite']):
    ax.text(i, v + 0.005, f'{v:.3f}', ha='center', fontsize=8.5, color=TICK)

# Tier annotations
ax.axhspan(rank_df['Composite'].max() * 0.80, rank_df['Composite'].max() * 1.05,
           alpha=0.08, color=C1, label='Tier 1')
ax.axhspan(rank_df['Composite'].max() * 0.50, rank_df['Composite'].max() * 0.80,
           alpha=0.06, color=C2, label='Tier 2')
ax.legend(fontsize=9)

plt.tight_layout(); plt.show()

print('\n─── Feature Ranking Table ────────────────────────────────────────────')
display(rank_df[['Feature','|Pearson|','|Spearman|','|Kendall|','Mutual Info','RF Importance','Composite']]
        .style.bar(subset=['Composite'], color=C1)
        .format({'|Pearson|':'{:.4f}','|Spearman|':'{:.4f}','|Kendall|':'{:.4f}',
                 'Mutual Info':'{:.5f}','RF Importance':'{:.5f}','Composite':'{:.4f}'}))
````

## 15 · EDA Findings, Feature Selection & Model Recommendation

### 15.1 — Key Dataset Findings

| Finding | Detail |
|---|---|
| **Single-repo dataset** | All 102 rows come from `plankanban/planka`. YAML-derived features (steps, line count, action count) are **identical across all workflows** — they carry between-workflow signal only, not within-workflow signal. |
| **Bimodal target (pre-removal)** | Two clusters: short (~0.5–2.1 min) and long (~28–31 min). The long cluster corresponds to `Build and Push Docker *` workflows. |
| **24 outliers removed** | IQR fence at 3.77 min. Removed rows are the `Build and Push Docker Image / Nightly Image` workflows. |
| **Zero-variance features (post-removal)** | `yaml_depth`, `if_condition_count`, `timeout_minutes`, `needs_dependencies_count`, `job_count`, `matrix_permutations`, `matrix_dimensions`, `uses_matrix_strategy`, `fail_fast`, `code_complexity` — all constant in the clean dataset. They must be **dropped**. |
| **Primary predictor: workflow_name** | Ordinal-encoded by mean duration, Spearman ρ ≈ 0.98, MI ≈ 1.81. It encodes the full YAML structure compactly. |
| **OS label has zero effect** | Every workflow achieves the same mean duration on both ubuntu and self-hosted runners in this dataset. |
| **Non-normal target** | Shapiro-Wilk rejects normality after outlier removal → apply `log1p` transform when training linear models. |

### 15.3 — Model Recommendation

**Primary: XGBoost Regressor (or LightGBM)**

| Criterion | Justification |
|---|---|
| Non-normal, skewed target | `log1p`-transform target; XGBoost on log scale → RMSLE metric |
| Non-linear relationships | Spearman > Pearson for key features — tree splits capture threshold effects naturally |
| Mixed feature types | Native support for numeric + binary + categorical (with ordinal encoding) |
| Small dataset (78–102 rows) | L1/L2 regularisation (`reg_alpha`, `reg_lambda`) prevents overfitting |
| Cold-start prediction | Must generalise to unseen repos → regularised models outperform memorisation models |
| Interpretability needed | SHAP values provide per-prediction feature attribution |

**Validation strategy — Group K-Fold by `repo_name`:**
```python
from sklearn.model_selection import GroupKFold
# Groups = repo_name ensures test folds contain UNSEEN repositories,
# simulating the cold-start scenario the model will face in production.
gkf = GroupKFold(n_splits=5)
for train_idx, test_idx in gkf.split(X, y, groups=df['repo_name']):
    ...
```

**Model stack:**

| Priority | Model | Purpose |
|---|---|---|
| 1 ⭐ | `XGBRegressor` / `LGBMRegressor` | Primary — handles all identified data characteristics |
| 2 | `RandomForestRegressor` | Robust baseline; compare against XGBoost |
| 3 | `Ridge` on `log1p(y)` | Linear baseline — quantifies how much non-linearity matters |

**Target transform + metric:**
```python
import numpy as np
from sklearn.metrics import mean_squared_error

y_train = np.log1p(df['duration_minutes'])     # transform
y_pred  = np.expm1(model.predict(X_test))      # inverse-transform predictions

# Primary metric: RMSLE — symmetric penalty on relative errors
rmsle = np.sqrt(mean_squared_error(np.log1p(y_true), np.log1p(y_pred)))
```
